# Пример на Seaborn

Heatmap средней оценки фильмов за последние 15 лет в топовых(из прошлого пункта) странах

```python
import seaborn as sns
```

```python
top_countries_df = df.loc[df['country'].isin(top_countries.index.values) & df['title_year'].isin(array(range(2001,2017)))]
top_countries_df
```

```python
country_year_imdbscore = top_countries_df.pivot_table(index='country',columns='title_year',\
                                        values='imdb_score',aggfunc=mean).fillna(0).applymap(float)
sns.heatmap(country_year_imdbscore, annot=True, fmt='.1f',linewidths=5.00)
```

# Пример на plotly

Средний бюджет фильмов по годам с количеством фильмов выпущенных в том или ином году

```python
from plotly.offline import init_notebook_mode, iplot
import plotly
import plotly.graph_objs as go

init_notebook_mode(connected=True)
```

```python
budget_by_years = df.groupby('title_year')[['budget']].mean()\
                    .join(df.groupby('title_year')[['movie_title']].count())
budget_by_years.columns = ['Budget', 'Number_of_movies']
budget_by_years
```

```python
trace0 = go.Scatter(
    x=budget_by_years.index,
    y=budget_by_years.Budget,
    name='Budget'
)

trace1 = go.Scatter(
    x=budget_by_years.index,
    y=budget_by_years.Number_of_movies,
    name='Number of movies'
)

data = [trace0, trace1]

fig = go.Figure(data=data)

iplot(fig) #, show_link=False)
```
